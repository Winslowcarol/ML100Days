# 範例重點

如何使用亂數, 資料集來操作


```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np

#畫出完整的 cos 圖形
'''
作業

'''
x = np.arange(-360,360)
y = np.cos(x*np.pi/90)
plt.xlim(-400,400)
plt.plot(x,y)
plt.show() 
plt.savefig("HW_D17.png",dpi=300,format="png")
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

#給定散點圖顏色
'''
#作業

#方法不只有一種

'''
color=np.arctan2(X,Y)
plt.scatter(X,Y, c=color) 
plt.xlim(-1.5,1.5) 
plt.ylim(-1.5,1.5)
plt.show()
```
